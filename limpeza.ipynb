### Resetando o Dataframe

```python
df = pd.read_csv("ocorrencia_2010_2020.csv", sep=";", parse_dates=["ocorrencia_dia"], dayfirst=True)
df.head(5)
```

# Limpeza dos Dados

Observa no arquivo CSV no Libre os dados que precisam ser limpos

```python
df.head(5)
```

```python
#### Limpando as linhas baseados em filtros
df.loc[df.ocorrencia_aerodromo == "****", ["ocorrencia_aerodromo"]] = pd.NA # atribui valores nao disponiveis em uma celula
# o pandas tem funcoes para trabalhar com dados nao informados
```

```python
df.head(3)
```

#### Função replace
Permite fazer alterações no dataset como um todo. procura alguns valores e altera por outros valores

```python
df.replace(["**", "###!", "####","****", "*****", "NULL"], pd.NA, inplace=True)
```

```python
df.head()
```

```python
#Quantos NA's eu tenho no dataframe?
#o pandas tem instruções para dados nao informados
df.isna().sum() # somar quantos dados nao informados por coluna
```

```python
df.isnull().sum()
```

```python
#### Colocar valores nesses dados nao informados
df.fillna(0, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

#### Alterando alguns dados nao informados

```python
df.replace([10], pd.NA, inplace=True)
```

```python
df.isnull().sum()
```

```python
# mudando valores na de uma coluna apenas
df.fillna(value={"total_recomendacoes": 0}, inplace=True)
```

```python
df.isnull().sum()
```

 # Criando e excluindo colunas

```python
# Criando coluna
df["total_recomendacoes_bkp"] = df.total_recomendacoes
```

```python
df.head(4)
```

```python
#excluindo colunas
df.drop(["total_recomendacoes_bkp"], axis=1, inplace=True) # modificando para o eixo da coluna
```

```python
df.head(3)
```

# Excluir os valores nao informados

```python
# Exclui todos os valores nao informados
df.dropna() #exclui a linha toda onde há na
```

```python
# exclui somente na coluna onde há na
#qual coluna procurar o na
df.dropna(subset=["ocorrencia_uf"])
```

# Valores duplicados

```python
df.drop_duplicates(inplace=True) # exclui valores duplicados
```
